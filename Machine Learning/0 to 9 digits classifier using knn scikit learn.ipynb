```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt

```

```python
digits = load_digits()
```

```python
print(digits)
```

```python
print(digits.target)
```

```python
dir(digits)
```

```python
digits.data
```

```python
df = pd.DataFrame(digits.data, digits.target, columns = digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df.head(15)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
X_train,  X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3, random_state=10)
```

```python
knn = KNeighborsClassifier(n_neighbors= 3)
```

```python
# len(X_train)
print(X_train.shape)
```

```python
len(y_test)
len(X_test)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_prediction = knn.predict(X_test)
cnm = confusion_matrix(y_test, y_prediction)
cnm
```

```python
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cnm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_prediction))
```
